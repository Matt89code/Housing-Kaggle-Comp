# Load the dataset

```python
traindf = pd.read_csv ('data/train.csv')
traindf.head()
```

```python
y = traindf['SalePrice']
y
```

```python
traindf.shape
```

```python
traindf = traindf.drop(columns = ['Id','SalePrice'])
```

```python
traindf.info()
```

# Data Preprocessing

## check columns quality and values

```python
catcol = [cols for cols in traindf.columns if traindf[cols].dtypes == 'object']
```

```python
numcol = [cols for cols in traindf.columns if (traindf[cols].dtypes ==  'int64') | (traindf[cols].dtypes == 'float64')]
```

```python
print(len(catcol))
print(len(numcol))
```

```python
traindf
```

```python
feat_categorical_nunique = traindf[catcol].nunique()
feat_categorical_nunique
```

## check histogram plot for unique values as some columns have such little varients! 

```python
plt.hist(feat_categorical_nunique, bins=20)
```

```python
feat_categorical_small = list(feat_categorical_nunique[feat_categorical_nunique < 7].index)
feat_categorical_small
```

# ML Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder as OHE
from sklearn.compose import ColumnTransformer    
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
    
numcols  = Pipeline([
    ('imputer', SimpleImputer(strategy = 'mean')),
    ('MinMaxScale', MinMaxScaler())])

catcols  = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('OHE', OHE(handle_unknown='ignore', sparse = False))])


preproc_baseline = ColumnTransformer([
    ('num_tr', numcols, numcol),
    ('cat_tr', catcols, feat_categorical_small) ]) 

preproc_baseline 
```

## adding pipeline and decision tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import make_pipeline

model =  DecisionTreeRegressor(max_depth=2, random_state=2)


pipe_baseline = make_pipeline(preproc_baseline, model)
pipe_baseline.fit(traindf,y)
```

## cross validation - using CV = 5 

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_log_error

rmsle = make_scorer(mean_squared_log_error)

cv_results = cross_val_score(pipe_baseline, traindf, y, cv=5, scoring=rmsle)
score_baseline = cv_results.mean()
score_baseline
```

# Predicting the y

```python
testdf = pd.read_csv ('data/test.csv')
testdf.head()
```

```python
y_pred_baseline = pipe_baseline.predict(testdf)
```

```python
y_pred_baseline
```

```python
df = pd.DataFrame()
df['Id'] = testdf['Id']
df['SalePrice'] = y_pred_baseline
df
```

```python
df.set_index('Id', inplace=True)
df
```
